```python
#Function to obtain MLS IDs from Redfin

MLS_IDs = []

def Find_MLS_ID(Property):
    driver.get("https://www.redfin.com/")
    time.sleep(1)
    search = driver.find_element_by_id("search-box-input")
    try:
        search.send_keys(Property)
        search.send_keys(Keys.RETURN)
        time.sleep(3)
        MLS_ID_element = driver.find_element_by_xpath('/html/body/div[1]/div[11]/div[2]/div[5]/section/div/div/div/div/div/div[4]/div[6]/span[2]')
        MLS_ID = MLS_ID_element.text
        MLS_IDs.append(MLS_ID)
    except:
        pass
```

```python
#Create Empty Data Frame to store all scraped data from MLS 

PropertyData = {}
PropertyData["Property_Name"] = []
PropertyData["List_Price"] = []
PropertyData["Sale_Price"] = []
PropertyData["DOM"] = []
PropertyData["Purchaser"] = []
PropertyData["Seller"] = []
PropertyData["Recording_Date"] = []
PropertyData["Sale_Date"] = []
df = pd.DataFrame(PropertyData)
```

```python
#Function to scrape Property Data From MLS

def MLS_Scrape(MLS_ID):
    
    #Look up Property in MLS based off MLS ID
    driver.get("https://matrix.crmls.org/Matrix/Default.aspx?c=AAEAAAD*****AQAAAAAAAAARAQAAAEQAAAAGAgAAAAQ4NDMzDUAGAwAAAAVwam7Cpw0CCw))&f=")
    try:
        continue_button = driver.find_element_by_link_text("Continue")
        continue_button.click()
        MLS_searchbar = driver.find_element_by_name("ctl02$m_ucSpeedBar$m_tbSpeedBar")
    except:
        try:
            MLS_searchbar = driver.find_element_by_name("ctl01$m_ucSpeedBar$m_tbSpeedBar")
        except:
            MLS_searchbar = driver.find_element_by_name("ctl02$m_ucSpeedBar$m_tbSpeedBar")   
    MLS_searchbar.send_keys(MLS_ID)
    MLS_searchbar.send_keys(Keys.RETURN)
    

    #Click Property Page
    time.sleep(1)
    Property_Page = driver.find_element_by_link_text(MLS_ID)
    Property_Page.click()
    
    
    #Establish a new dictonary row to add property attributes to which will then be added to the dataframe
    new_row = {"Sale_Price":[],"List_Price":[],"DOM":[],"Property_Name":[],"Purchaser":[],"Seller":[],"Recording_Date":[],"Sale_Date":[]}
    
    
    #Function to locate and scrape the many attributes that we're interested in for each property
    def LocateAttribute(AttributeName, xpath):
        nonlocal new_row
        try:
            element = driver.find_element_by_xpath(xpath)
            AttributeValue = element.text
            new_row[AttributeName].append(AttributeValue)
        except:
            new_row[AttributeName].append("Error/Null")

    
#LISTING PAGE ATTRIBUTES 
    
    #Go to listing page
    time.sleep(1)
    Listing_Page = driver.find_element_by_link_text("Listing")
    Listing_Page.click()
    time.sleep(1)
    #List Price Attribute
    LocateAttribute("List_Price", "/html/body/form/div[3]/div[6]/table/tbody/tr/td/div[2]/div[3]/div[3]/div/div/div[1]/div/div/div/div/div[3]/div/span/div[2]/div/table/tbody/tr/td/table/tbody/tr/td/table/tbody/tr[2]/td/span/div/table/tbody/tr/td/table/tbody/tr[3]/td/table[9]/tbody/tr[16]/td/table[9]/tbody/tr[1]/td/table[2]/tbody/tr/td[1]/table/tbody/tr[2]/td[1]/span[3]")
    #Sale Price Attribute
    LocateAttribute("Sale_Price", "/html/body/form/div[3]/div[6]/table/tbody/tr/td/div[2]/div[3]/div[3]/div/div/div[1]/div/div/div/div/div[3]/div/span/div[2]/div/table/tbody/tr/td/table/tbody/tr/td/table/tbody/tr[2]/td/span/div/table/tbody/tr/td/table/tbody/tr[3]/td/table[9]/tbody/tr[16]/td/table[9]/tbody/tr[1]/td/table[2]/tbody/tr/td[1]/table/tbody/tr[1]/td[1]/span[3]")
    #DOM Attribute
    LocateAttribute("DOM", "/html/body/form/div[3]/div[6]/table/tbody/tr/td/div[2]/div[3]/div[3]/div/div/div[1]/div/div/div/div/div[3]/div/span/div[2]/div/table/tbody/tr/td/table/tbody/tr/td/table/tbody/tr[2]/td/span/div/table/tbody/tr/td/table/tbody/tr[3]/td/table[9]/tbody/tr[16]/td/table[9]/tbody/tr[1]/td/table[2]/tbody/tr/td[1]/table/tbody/tr[5]/td[1]/span[3]/a[1]")
    


#TAX PAGE ATTRIBUTES
    
    #Go to tax page
    Tax_Page = driver.find_element_by_link_text("Tax")
    Tax_Page.click()
    time.sleep(1)
    #Property Name Attribute
    LocateAttribute("Property_Name", "/html/body/form/div[3]/div[6]/table/tbody/tr/td/div[2]/div[3]/div[3]/div/div/div[1]/div/div/div/div/div[1]/div/span[2]")
    #Purchaser Attribute
    LocateAttribute("Purchaser", "/html/body/form/div[3]/div[6]/table/tbody/tr/td/div[2]/div[3]/div[3]/div/div/div[1]/div/div/div/div/div[3]/div/span/div[2]/div/table/tbody/tr/td/table/tbody/tr/td/span/div/div/div/div/div[2]/div/div/div[1]/div/div[2]/span")
    #Seller Attribute
    LocateAttribute("Seller", "/html/body/form/div[3]/div[6]/table/tbody/tr/td/div[2]/div[3]/div[3]/div/div/div[1]/div/div/div/div/div[3]/div/span/div[2]/div/table/tbody/tr/td/table/tbody/tr/td/span/div/div/div/div/div[18]/div/div/div[5]/div/div[2]/span")
    #Recording Date Attribute
    LocateAttribute("Recording_Date", "/html/body/form/div[3]/div[6]/table/tbody/tr/td/div[2]/div[3]/div[3]/div/div/div[1]/div/div/div/div/div[3]/div/span/div[2]/div/table/tbody/tr/td/table/tbody/tr/td/span/div/div/div/div/div[18]/div/div/div[1]/div/div[2]/span")
    #Sale Date Attribute
    LocateAttribute("Sale_Date", "/html/body/form/div[3]/div[6]/table/tbody/tr/td/div[2]/div[3]/div[3]/div/div/div[1]/div/div/div/div/div[3]/div/span/div[2]/div/table/tbody/tr/td/table/tbody/tr/td/span/div/div/div/div/div[18]/div/div/div[2]/div/div[2]/span")
    
    
    #Add property data to df
    global df
    df = df.append(new_row, ignore_index=True)
    
```

```python
#Establish driver path
Path = "/Applications/chromedriver"
driver = webdriver.Chrome(Path)


#Login to South Bay MLS
driver.get("https://dashboard.cl.crmls.org/layouts")
username_input = driver.find_element_by_id("clareity")
password_input = driver.find_element_by_id("security")
username_input.send_keys(My_Username)
password_input.click()
password_input.send_keys(My_Password, Keys.RETURN)


for property in property_list:
    Find_MLS_ID(property)


for MLS_ID in MLS_IDs:
    MLS_Scrape(MLS_ID)


#Create record of MLS IDs used in scrape
MLS_IDs_Record = MLS_IDs


driver.quit()

```

```python
#Remove the brackets that are added to data when appended to dataframe

def RemoveBrackets(AttributeName):
    Array = df[AttributeName].tolist()
    
    cleanArray = []
    
    for x in Array:
        #Remove brackets
        y = str(x)[1:-1]
        #Remove ' ' that are added when converted to string
        z = y[1:-1]
        cleanArray.append(z)
    
    df.drop(AttributeName, axis=1)
    df[AttributeName] = cleanArray

for column in df:
    RemoveBrackets(column)
```
